```python
import pandas as pd # It is used to work on structured data
import matplotlib.pyplot as plt # It is used to visualize the data
import seaborn as sns # It is used to do visualization of Regression very neatly
```

```python
df= pd.read_csv('temperatures.csv')
```

```python
df
# To see first five lines we can write as --> df.head()
```

```python
# Input Data
x=df['YEAR']

# Output Data
y= df['ANNUAL']
```

```python
# plt.figure(figsize=(16,9)) # To give the size to the plot
plt.title('Temperatue Plot of INDIA') # To give the Title to the Plot
plt.xlabel('YEAR') # To give the X-axis name
plt.ylabel('ANNUAL Average Temperature') # To give the Y-axis name
plt.scatter(x,y) # To give the type of the plot/graph
```

```python
# In order to make the Regression Model we have to give Two Dimensional Data means rows and columns both
# But as you see here it has only rows and not columns
# Note it is the Requirement of Python Library and not of Machine
x.shape
```

```python
x=x.values # To convert the values in an Array
```

```python
x
```

```python
x=x.reshape(117,1) # Here we can give the requires no. of rows and columns using "Reshape"
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression # To import a class named "LinearRegression" from "scikit learn"
```

```python
regressor= LinearRegression() # To create object of the class "LineearRegression"
```

```python
regressor.fit(x,y) # To train our algorithm we use "fit()"
```

```python
regressor.coef_ # To get the "Slope" of regression i.e. "m" of "y=mx+c"
```

```python
regressor.intercept_ # To get the value "c" of "y=mx+c"
```

```python
regressor.predict([[2024]]) # To predict the values. Here we are predicting the Average Temperature of the Year 2024.(Note:2 Square Brackets)
```

```python
regressor.predict([[2035]])
```

```python
regressor.predict([[2070]])
```

```python
predicted= regressor.predict(x) # To get the Predicted Values from the Algorithm
```

```python
predicted
```

```python
y # It is the actual value
```

```python
# Here we use numpy to perform Mathematical Operations
import numpy as np
```

```python
# To caluculate Mean Absolute Error
np.mean(abs(y - predicted))
```

```python
# To find the MAE, we have Prebuilt Function as
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,predicted) # Here we have to pass the Actual and Predicted values
```

```python
# To find Mean Squared Error(MSE)
np.mean((y-predicted)**2) # Here we took Square of the answer
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,predicted) # Here we are using the function for finding the MSE
```

```python
# To find the value of R-Square Metrics i.e. R-Squared Error
from sklearn.metrics import r2_score # r2_score is used to find out the linearity in the data
r2_score(y,predicted)
```

```python
# We can also use following to get the value of R-Squared Error
regressor.score(x,y)
```

```python
# plt.figure(figsize=(16,9)) # To give the size to the plot
plt.title('Temperatue Plot of INDIA') # To give the Title to the Plot
plt.xlabel('YEAR') # To give the X-axis name
plt.ylabel('ANNUAL Average Temperature') # To give the Y-axis name
plt.scatter(x,y, label='actual', color='r', marker='.') # To give the type of the plot/graph. Here we used "Marker" as "." to get the small dots
plt.plot(x,predicted, label='predicted', color='g')
plt.legend() # This is used to get the values with their colors in the form of a "Box"
```
